```python
# 9) seu código aqui
df['asfixia'] = np.where(
    df['APGAR5'] <=3 , 'asfixia severa',
    np.where(df['APGAR5'] <= 5, 'asfixia moderada',
             np.where(df['APGAR5'] <= 7, 'asfixia leve', 'normal')))

df['asfixia'].value_counts()
```

```python
df.loc[df['APGAR5'] <= 3, 'asfixia2'] = 'asfixia severa'
df.loc[(df['APGAR5'] == 4) | (df['APGAR5'] == 5), 'asfixia2'] = 'asfixia moderada'
df.loc[(df['APGAR5'] == 6) | (df['APGAR5'] == 7), 'asfixia2'] = 'asfixia leve'
df.loc[(df['APGAR5'] >= 8), 'asfixia2'] = 'normal'

df['asfixia2'].value_counts()
```
